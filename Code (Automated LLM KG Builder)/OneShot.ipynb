```python
from langchain_community.llms import Ollama
from langchain_core.output_parsers import JsonOutputParser
from langchain.output_parsers import YamlOutputParser
```

```python
import pandas as pd
import numpy as np
```

```python
llm = Ollama(model="gemma:7b")
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a ontologist. You build knowledge graph from plain input by using example Knowledge Graph whis is Cypher format. Create Knowledge Graph in Cypher Script"),
    ("system", "{example}"),
    ("user", "{input}")
])

chain = prompt | llm
```

```python
example_kg = '''
CREATE (beyonce:Person {name: "Beyonce"})
CREATE (boyfriend:Person {name: "Not Mentioned"})
CREATE (foxxy:Character {name: "Foxxy Cleopatra”,  date: “July 2002”})

CREATE (boyfriend)-[:LEFT]->(beyonce)
CREATE (beyonce)-[: ACT]->(austinpowers)
CREATE (beyonce)-[: ACT_AS]->(foxxy)
'''
```

```python
contexts = pd.read_csv('/Users/mertakcay/Projects/KnowledgeGraphBuilder/context.csv')
for context in contexts['context'].values:
    entities = chain.invoke({"input": context, "example": example_kg})
    print(entities)
    print('*****************')
    # for sample in entities.split('\n'):
    #     if '-' in sample:
    #         entity_list.append(sample.replace('- ',''))
```
