```python
from langchain_community.llms import Ollama
from langchain_core.output_parsers import JsonOutputParser
from langchain.output_parsers import YamlOutputParser
```

```python
import pandas as pd
import numpy as np
```

```python
llm = Ollama(model="gemma:7b")
```

```python
from langchain_core.prompts import ChatPromptTemplate
import re

def extract_cypher_scripts(text):
    cypher_list = list()
    for line in text.split("\n"):
        if 'CREATE' in line or 'MATCH' in line or 'CONNECT' in line:
            cypher_list.append(line)
            
    return cypher_list
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a ontologist. You build knowledge graph from plain input by using example Knowledge Graph whis is Cypher format. Create Knowledge Graph in Cypher Script"),
    ("system", "{example}"),
    ("user", "{input}")
])

chain = prompt | llm
```
